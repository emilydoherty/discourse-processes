## Individual-Level Results

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel, ttest_ind
```

```python

df = pd.read_excel('/Users/emilydoherty/Library/CloudStorage/OneDrive-UCB-O365/Emily_Papers/iSAT_discoursepaper2023/Other Measures/AICLsurvey_long.xlsx', sheet_name='Sheet1')  # Replace 'Sheet1' with your actual sheet name if different
df_filtered= df[df['Round'].isin([2, 4])]
drop_columns = ['Wason_1_Attempts', 'Wason_2_Attempts', 'Wason_Overall', 'Treasure_Ct', 'BoardGameSuccessRC' ,'SpheroTaskSuccessRC']
df_filtered.drop(columns=drop_columns,inplace=True)
# List of survey measures
survey_measures = ['TLXMentalDemand', 'TLXRush', 'TLXSuccessRC', 'SAMPleasure', 
                   'SAMArousal', 'SocialLoafAvg', 'SLPosIntAvg', 'SocCohAvg', 
                   'CogTrust', 'AffTrust', 'MonitoringRC', 'PsychSafetyAvg']
```

```python
## Test for the effect of task 

# Iterate through each measure
for measure in survey_measures:
    makecode_data = df[df['Round'] == 4][measure]
    weights_data = df[df['Round'] == 2][measure]

    # Perform a paired t-test
    t_statistic, p_value = ttest_rel(makecode_data, weights_data) #paired t-test 

    # Print the results
    print(f'Measure: {measure}')
    print(f'T-statistic: {t_statistic}')
    print(f'P-value: {p_value}')

    if p_value < 0.05:
        print(f"There is a significant difference between 'makecode' and 'weights' rounds for {measure}.")
    else:
        print(f"There is no significant difference between 'makecode' and 'weights' rounds for {measure}.")
    print('\n')

```

```python
# Iterate through each measure
for measure in survey_measures:
    print(f'Measure: {measure}')

    # Get data for each group size
    group_data_2 = df[df['Group_Size'] == 2][measure]
    group_data_3 = df[df['Group_Size'] == 3][measure]

    # Perform an independent t-test
    t_statistic, p_value = ttest_ind(group_data_2, group_data_3) #independent ttest for 2 seperate groups 

    # Print the results
    print(f'T-statistic: {t_statistic}')
    print(f'P-value: {p_value}')

    if p_value < 0.05:
        print(f"There is a significant difference for {measure} between Group Size 2 and 3.")
    else:
        print(f"There is no significant difference for {measure} between Group Size 2 and 3.")
    print('\n')

```

```python
###NEED TO FIX NANs here 
# Specify the measure of interest
measure_of_interest = 'MakeCodeOverallSuccessRC'

# Get data for each group size
group_data_2 = df[df['Group_Size'] == 2][measure_of_interest]
group_data_3 = df[df['Group_Size'] == 3][measure_of_interest]

# Perform an independent t-test
t_statistic, p_value = ttest_ind(group_data_2, group_data_3)

# Print the results
print(f'Measure: {measure_of_interest}')
print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')

if p_value < 0.05:
    print(f"There is a significant difference for {measure_of_interest} between Group Size 2 and 3.")
else:
    print(f"There is no significant difference for {measure_of_interest} between Group Size 2 and 3.")

```
